```python
def flatten_json(json):
    dict1 = {}
    
    def flatten(i, name=""):
        if type(i) is dict:
            for a in i:
                flatten(i[a], name + a + '_')
        else:
            dict1[name[:-1]] = i
    
    flatten(json)
    return dict1
```

```python
#Data Frame

dados = pd.DataFrame.from_dict(flatten_json(parsed_data),orient="index")
dados
```
